9.What is type casting in Python?
ans.Type casting in Python, also known as type conversion, is the process of converting a value from one data type to another. This is often necessary when you need to perform operations that require specific data types, or when you need to store data in a different format.

**There are two main types of type casting in Python:**
->Implicit Type Conversion (Coercion):
This occurs automatically when Python converts a data type to another without explicit user intervention. This typically happens in expressions where different data types are involved, and Python decides how to handle the conversion to avoid data loss or errors. For example, when an integer is added to a float, the integer is implicitly converted to a float to ensure the result is a float.
->Explicit Type Conversion (Type Casting):
This involves manually converting a data type using built-in functions or constructors. This allows you to explicitly control the conversion process and ensures that the data is converted to the desired type. Common functions used for explicit type casting include:
.int(): Converts to an integer.
.float(): Converts to a floating-point number.
.str(): Converts to a string.
.list(): Converts to a list.
.tuple(): Converts to a tuple.

**While Loops:**

9.Write a Python program to convert user input from string to integer, float, and boolean types.

```python
user_input = input("Enter a value: ")

# Convert to integer
if user_input.isdigit():
    int_value = int(user_input)
    print(f"Integer value: {int_value}")
else:
    print("Cannot convert to integer.")

# Convert to float
parts = user_input.split('.')
if len(parts) <= 2 and all(part.isdigit() for part in parts):
    float_value = float(user_input)
    print(f"Float value: {float_value}")
else:
    print("Cannot convert to float.")

# Convert to Boolean
if user_input.lower() == "true":
    bool_value = True
elif user_input.lower() == "false":
    bool_value = False
else:
    bool_value = "Unknown"
print(f"Boolean value: {bool_value}")
```

10. Write code to demonstrate type casting with list elements.

```python
numbers_str = ["1", "2", "3"]
numbers_int = [int(x) for x in numbers_str]
print(numbers_int)  # Output: [1, 2, 3]

numbers_float = [float(x) for x in numbers_int]
print(numbers_float)  # Output: [1.0, 2.0, 3.0]

numbers_str_again = [str(x) for x in numbers_float]
print(numbers_str_again)  # Output: ['1.0', '2.0', '3.0']
```

15.Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial of", num, "is:", fact)
```
